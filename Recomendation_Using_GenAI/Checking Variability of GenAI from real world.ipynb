### Recommended Nutritional Content per 100 g for Meals in Type 2 Diabetes Patients

| Nutrient      | Recommended Range*                 |
|---------------|------------------------------------|
| **Calories**  | 100200 kcal                      |
| **Protein**   | 510 g                            |
| **Carbohydrates** | 1015 g (preferably low GI carbs) |
| **Fiber**     | 36 g                             |
| **Fat**       | 26 g (focus on unsaturated fats) |
| **Sugar**     | < 5 g (preferably natural sugars) |

*These values are general guidelines and should be adapted to individual needs under a dietitian's supervision.

```python
requirements_data = {
    "Calories (kcal)": "50-150",
    "Protein (g)": "4-10",
    "Carbohydrates (g)": "10-20",
    "Fiber (g)": "2-6",
    "Fat (g)": "1-8",
    "Sugar (g)": "<5"
}
```

```python
requirements_data
```

```python
# Define the real requirements as midpoints of the ranges
requirements_data = {
    "Calories (kcal)": (50 + 150) / 2,  # midpoint of range
    "Protein (g)": (4 + 10) / 2,
    "Carbohydrates (g)": (10 + 20) / 2,
    "Fiber (g)": (2 + 6) / 2,
    "Fat (g)": (1 + 8) / 2,
    "Sugar (g)": 5  # upper bound used
}

# AI model averages
llama_avg = {
    "Calories (kcal)": 309.52,
    "Protein (g)": 24.76,
    "Carbohydrates (g)": 23.81,
    "Fiber (g)": 4.14,
    "Fat (g)": 11.90,
    "Sugar (g)": 10.24
}

gemini_avg = {
    "Calories (kcal)": 249.76,
    "Protein (g)": 19.33,
    "Carbohydrates (g)": 21.24,
    "Fiber (g)": 5.00,
    "Fat (g)": 8.64,
    "Sugar (g)": 4.93
}

# Convert to DataFrame for easier calculations
real = pd.Series(requirements_data)
llama = pd.Series(llama_avg)
gemini = pd.Series(gemini_avg)

# Calculate metrics
def calculate_metrics(predicted, real):
    absolute_error = np.abs(predicted - real)
    percentage_error = (absolute_error / real) * 100
    mse = np.mean((predicted - real) ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(absolute_error)
    return pd.DataFrame({
        "Absolute Error": absolute_error,
        "Percentage Error (%)": percentage_error,
        "MSE": mse,
        "RMSE": rmse,
        "MAE": mae
    })

# Metrics for LLaMA and Gemini
llama_metrics = calculate_metrics(llama, real)
gemini_metrics = calculate_metrics(gemini, real)

llama_metrics, gemini_metrics
```
